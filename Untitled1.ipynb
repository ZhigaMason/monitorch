```python
import torch
import torch.nn as nn

from monitorch.inspector import PyTorchInspector
from monitorch.lens import (
    ParameterGradientActivation,
    ParameterGradientGeometry,
    ParameterNorm,
    LossMetrics
)
```

```python
from torchvision.datasets import CIFAR10
from torch.utils.data import DataLoader
from torchvision.transforms import v2

transform = v2.Compose([v2.ToImage(), v2.ToDtype(torch.float32, scale=True)])
ds_train = CIFAR10('./tmp_data/', train=True, download=True, transform=transform)
ds_valid = CIFAR10('./tmp_data/', train=False, download=True, transform=transform)

train_data = DataLoader(
    ds_train,
    batch_size=32,
    shuffle=True
)

valid_data = DataLoader(
    ds_valid,
    batch_size=32
)
```

```python
from collections import OrderedDict
from torch.nn.parameter import Parameter


class MyANN(nn.Module):

    def __init__(self, input_dim, n_classes):
        super().__init__()
        self.weight = Parameter(torch.normal(0.0, 1.0, (input_dim, 256)))
        self.bias = Parameter(torch.rand(256))
        
        self.sub = nn.Sequential(OrderedDict([
            ('lin1', nn.Linear(256, 256)),
            ('relu1', nn.ReLU()),

            ('lin2', nn.Linear(256, 256)),
            ('relu2', nn.ReLU()),

            ('lin_out', nn.Linear(256, 10)),
            ('output', nn.Softmax(dim=1))
        ]))

    def forward(self, X):
        act = torch.abs(X.flatten(start_dim=1) @ self.weight + self.bias)
        return self.sub(act)

model = MyANN(3 * 32 * 32, 10)

loss_fn = nn.CrossEntropyLoss()

inspector = PyTorchInspector(
    lenses= [
        LossMetrics(loss_fn=loss_fn),
        ParameterGradientGeometry(),
        ParameterGradientActivation(),
        ParameterNorm(),
    ],
    module=model
)

model
```

```python
torch.set_num_threads(4)
```

```python
from tqdm.notebook import trange, tqdm
NEPOCHS = 2

optimizer = torch.optim.AdamW(
    model.parameters()
)

def train_one_epoch(model, optimizer, loss_fn, train_data):
    for imgs, labels in tqdm(train_data):
        optimizer.zero_grad()
        pred = model(imgs)
        loss = loss_fn(pred, labels)
        loss.backward()
        optimizer.step()

@torch.no_grad
def validate_one_epoch(model, loss_fn, valid_data):
    for imgs, labels in tqdm(valid_data):
        pred = model(imgs)
        loss = loss_fn(pred, labels)

for epoch in trange(NEPOCHS):

    train_one_epoch(model, optimizer, loss_fn, train_data)
    validate_one_epoch(model, loss_fn, valid_data)

    inspector.tick_epoch()

fig = inspector.visualizer.show_fig()
    
```
